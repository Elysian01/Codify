```python
import pandas as pd
```

```python
codebase = pd.read_json("codify_codebase.json")
codebase.head(10)
```

```python
%%time
intent = "null_imputation"
entity_name = "STATISTICS"
entity_value = "median"

# intent = "classify_eval"
# entity_name = "evaluation"
# entity_value = "classification"

result = []

intent_search_result = codebase.loc[codebase["intent"] == intent]
if not intent_search_result.empty:
    entity_name_search_result = intent_search_result.loc[codebase["entity_name"] == entity_name]
    if not entity_name_search_result.empty:
        entity_value_search_result = entity_name_search_result.loc[codebase["entity_value"] == entity_value]
        if not entity_name_search_result.empty:
            result.append(entity_value_search_result["code"].values)
        else:
            result.append(entity_name_search_result["code"].values)
    else:
        result.append(intent_search_result["code"].values)
    
print(result[0])
```
